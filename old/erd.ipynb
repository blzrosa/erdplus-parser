````python
important_features_nodes = []
irrelevant_features = ['position', 'isSelected', 'isConnectable', 'selected', 'dragging', 'measured']
for node in data_serialized['nodes']:
    node_relevant = {}
    for key, value in node.items():
        if key not in irrelevant_features:
            if key == 'data':
                new_data = {}
                for k, v in node['data'].items():
                    if k not in irrelevant_features:
                        new_data[k] = v
                node_relevant['data'] = new_data
            else:
                node_relevant[key] = value
    important_features_nodes.append(node_relevant)
````

````python
important_features_nodes
````

````python
unique_types = set()
for node in important_features_nodes:
    unique_types.add(node['type'])
print(list[unique_types])
````

````python
labels = [node for node in important_features_nodes if node['type'] == 'Label']
entities = [node for node in important_features_nodes if node['type'] == 'Entity']
attributes = [node for node in important_features_nodes if node['type'] == 'Attribute']
relationships = [node for node in important_features_nodes if node['type'] == 'Relationship']
````

Attribute:
